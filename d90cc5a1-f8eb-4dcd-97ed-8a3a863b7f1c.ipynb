# Paso 10

Apliquemos ahora todos los cambios a la lista de clientes. Para simplificar las cosas, te proporcionaremos una más corta.
Debes:
1. Eliminar todos los espacios iniciales y finales de los nombres, así como cualquier guion bajo.
2. Convertir todas las edades en números enteros.
3. Separar todos los nombres y apellidos en una sublista.

Guarda la lista modificada como una nueva lista llamada `users_clean` y muéstrala en la pantalla.


```python
users = [
    ['32415', ' mike_reed ', 32.0, ['ELECTRONICS', 'SPORT', 'BOOKS'], [894, 213, 173]],
    ['31980', 'kate morgan', 24.0, ['CLOTHES', 'BOOKS'], [439, 390]],
    ['32156', ' john doe ', 37.0, ['ELECTRONICS', 'HOME', 'FOOD'], [459, 120, 99]],
]

users_clean = [] #Dejamos la lisrta vacía porque procesaremos por individual a cada usuario.


# Procesa al primer usuario
user_name_1 = users[0][1].strip().replace("_", " ") #Se utiliza el 0 porque es la primera línea y Pyhton empieza a contar desde cero. Así también tenemos que el nombre se encuentra en la posición 1. Le indicamos al programa que queremos quitar los espacios iniciales y cambiar el guión bajo por un espacio.
user_age_1 = int(users[0][2]) #Transformamos a entero el valor 2 que corresponde a la edad. Seguimos en la línea 0.
user_name_1 = user_name_1.split() #Para separar el nombre llamamos a la función split.
users_clean.append([users[0][0], user_name_1, user_age_1, users[0][3], users[0][4]]) #Añadimos a users_clean lo que acabamos de hacer, con la adición de los otros elementos de la línea que no tuvimos que trabajar, en el orden en el que van.

# Procesa al segundo usuario
user_name_2 = users[1][1].strip()  #Se repite el procedimiento, pero aquí ya es la línea 1. Aquí no es necesario utilizar replace porque el nombre ya está separado por un espacio y no por algun otro caracter que necesitemos eliminar.
user_age_2 = int(users[1][2])
user_name_2 = user_name_2.split()
users_clean.append([users[1][0], user_name_2, user_age_2, users[1][3], users[1][4]])

# Procesa al tercer usuario
user_name_3 = users[2][1].strip() #Se repite el procedimiento, pero aquí ya es la línea 2. Tampoco es necesario utilizar replace para el nombre.
user_age_3 = int(users[2][2])
user_name_3 = user_name_3.split()
users_clean.append([users[2][0], user_name_3, user_age_3, users[2][3], users[2][4]])



print(users_clean)
```
